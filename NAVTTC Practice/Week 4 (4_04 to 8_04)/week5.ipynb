```python
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
kashti=sns.load_dataset('titanic')
kashti.head()
```

### Types of Analysis
- Univariate Anlaysis
- Bivariate
- Multivariate

### Univariate

```python
sns.displot(x='sex',data=kashti)
```

```python
sns.histplot(x='class',data=kashti)
```

```python
sns.barplot(data=kashti)
```

```python
sns.countplot(data=kashti)
```

### Bivariate

```python
kashti.head()
```

```python
sns.scatterplot(x='survived',y='who',data=kashti)
```

```python
sns.catplot(x='alive',y='survived',data=kashti)
```

```python
sns.boxplot(x='sex',y='age',data=kashti)
```

```python
sns.relplot(x='who',y='sex',data=kashti)
```
